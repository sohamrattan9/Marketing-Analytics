```python
import pandas as pd
data={"Customer Segment":["Sneakerhead", "Fitness Enthusiast", "Casual Shopper", "Gen Z Hype Buyer", "Lapsed customer", "Loyal athlete", "Budget buyer", "Brand ambassador", "Teen trend follower", "Eco Conscious buyer"],
'App Engagement Score':[95, 88, 40, 93, 15, 85, 30, 100, 77, 65],
'Avg Monthly Spend':[12000, 8500, 2500, 9200, 0, 7800, 1500, 18000, 5000, 4200]}
df=pd.DataFrame(data)
print(df)
def trigger_message(row):
  engagement=row["App Engagement Score"]
  spend=row['Avg Monthly Spend']
  if engagement >=85 and spend >= 8000: return "Exclusive Drop Access For You!"
  elif engagement >=85 and spend<8000: return "Upgrade Your Game- New Gear Awaits"
  elif engagement <50 and spend>=8000: return "We've Missed You = Offers are Back!"
  elif engagement <50 and spend<3000: return "Come back again - 20% off only for you!"
  else: return "Check out whats Trending this week"

df["trigger Message"]=df.apply(trigger_message, axis=1)
print(df[["Customer Segment", "trigger Message"]])
```
